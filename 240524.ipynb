#붓꽃 데이터 읽어와서 확인하기

```python
#붓꽃 데이터 분석
import pandas as pd
iris_data = pd.read_csv('iris.csv')   #데이터 파일 읽어오기
iris_data.info()  #데이터 기본 정보 출력(info)
```

```python
# 데이터의 기초 통계량 출력
iris_data.describe()
```

```python
# 품종별 개수 구하기
iris_data['species'].value_counts()
#iris_data.species.value_counts()
```

#붓꽃 데이터 전처리

```python
#결측치 확인하기
iris_data.isnull().sum()
```

```python
#중복 데이터 확인하기
iris_data[iris_data.duplicated()]
```

```python
#중복 데이터 모두 확인하기
idx=(iris_data['sepal_length'] == 5.8) & (iris_data['petal_width'] == 1.9)
iris_data.loc[idx, :]
```

```python
#중복 데이터 삭제하기
iris_data=iris_data.drop_duplicates()
iris_data
```

#붓꽃 데이터 그룹핑

```python
#품종(species) 열을 기준으로 합계 구하기
iris_data.groupby('species').sum()
```

```python
#품종(species) 열을 기준으로 평균 구하기
iris_data.groupby('species').mean()
```

#붓꽃 데이터 시각화

###막대 그래프 그리기

```python
#꽃받침 길이의 시각화
import matplotlib.pyplot as plt
plt.bar(iris_data.index[:10], iris_data.sepal_length[:10])
plt.xlabel('Data')
plt.ylabel('Sepal length (cm)')
plt.show()
```

```python
#붓꽃 종류별 꽃받침 길이의 평균에 대한 시각화

df2=iris_data.groupby('species').mean()
plt.bar(df2.index,df2['sepal_length'])
plt.title('species sepal length graph')
plt.xlabel('species')
plt.ylabel('sepal_lengths mean')
plt.show()
```
