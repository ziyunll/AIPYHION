###시본(Seaborn)

```python
#import matplotlib.pyplot as plt
import seaborn as sns
#import pandas as pd
iris_data=pd.read_csv('iris.csv')
iris_data.info()
```

```python
sns.scatterplot(x='petal_length',y='petal_width'
,hue='species',data= iris_data)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
#실습문제: 다음 데이터 프레임에 대해서 산점도를 그리시오.
data = pd.DataFrame({
'x': [1, 2, 3, 4, 5],
'y': [10, 20, 15, 25, 30],
'category': ['A', 'B', 'A', 'B', 'A']
})

sns.scatterplot(x='x',y='y'
,hue='category',data= data)
plt.show()
```

```python
#실습문제: 주어진 데이터프레임에서 "category" 열은 데이터 포인트의 범주를, "value" 열은 해당 데이터 포인트의 값을 나타냅니다. 주어진 데이터를 이용하여 seaborn의 countplot() 함수를 사용하여 다음을 수행하세요:

data = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})

sns.countplot(x='category', hue='category', data=data)
#plt.legend()
plt.show()
```

```python
#MISSION 3: 시본에서 제공하는 데이터인 ‘penguins’를 이용해서 그래프 그리기
#1 penguins 데이터를 가져온 후 펭귄 품종(species)별 몸무게(body_mass_g)의 평균을 그래프로 그리는데, 아래 그림과 같이 성별을 구분해서 그려보자.
data2 = pd.read_csv('penguins.csv')
sns.barplot(x='species',y='body_mass_g',hue='sex', data=data2)
plt.show()
```

```python
#2 penguins 데이터의 숫자값을 갖는 항목들끼리의 상관관계 그래프를 그려보자.
sns.pairplot(data2)
plt.show()
```

```python
#3 2번 미션의 결과에서 칼럼들간의 상관관계를 확인하고 관계가 있는 두 칼럼을 선택하여 산점도와 추세선이 함께 나타나는 그래프를 그려보자.

sns.regplot(x='flipper_length_mm', y='body_mass_g', data=data2)
plt.show()
```
